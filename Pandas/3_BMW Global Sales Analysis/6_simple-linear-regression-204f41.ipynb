# **Simple Linear Regression**

```python
# In this worksheet you will be introduced to simple linear regression. That is: regression with one dependent and one independednt variable. 
```

```python
# Step 1 : import library
import pandas as pd
```

```python
# Step 2 : import data note that we are saving our data as a salary dataframe (a type of object)
salary = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')

data = pd.read_csv('/kaggle/input/bmw-global-sales-analysis/BMW sales data (2010-2024) (1).csv')
```

```python
# Step 3 : define target (y) and features (X)
```

```python
salary.columns
data.columns
```

```python
y = data['Price_USD']
```

```python
X = data[['Mileage_KM']]
```

```python
# Step 4 : train test split 
# Check this step carefully to see what each part does.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 : select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Step 7 : predict model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Step 8 : model accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
