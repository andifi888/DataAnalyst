```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = 'cosmetics_sales_data.csv'
data = pd.read_csv(file)
```

# EDA

```python
print("\nShow first 5 rows:\n")
data.head()
```

```python
print("\nGeneral Info. about dataset:\n")
data.info()
```

```python
print("\nDescriptive Stats. for each column:\n")
data.describe()
```

## Unique Values

```python
for column in data.columns:
    unique_values = data[column].unique()
    print(f"Unique values in the columns '{column}':")
    print(unique_values)
```

```python
for country in data['Country'].unique():
    print(f"\nData for {country}:")
    print(data[data['Country'] == country])
```

```python
for country in data['Country'].unique():
    total = data[data['Country'] == country]['Amount ($)'].sum()
    print(f"{country}: {total:.2f} $")
```

```python
for country in data['Country'].unique():
    count = data[data['Country'] == country].shape[0]
    print(f"{country} have {count} sales")
```

```python
for col in data.columns:
    unique_count = data[col].nunique()
    print(f"Column '{col}': {unique_count} unique values")
```

## Empty values

```python
print("\nNULL values:\n")
print(data.isnull())
```

```python
print("\nNumber of null values for each column:\n")
print(data.isnull().sum())
```

```python
print("\nNumber of null values for each row:\n")
print(data[data.isnull().any(axis=1)])
```

## Graphs

```python
fig = plt.figure(figsize=(20, 16))

plt.subplot(1, 3, 1)
country_sales = data.groupby('Country')['Amount ($)'].sum().sort_values(ascending=False)
plt.bar(country_sales.index, country_sales.values, color='skyblue')
plt.title('Total Sales Amount by Country', fontsize=14, fontweight='bold')
plt.xlabel('Country')
plt.ylabel('Amount ($)')
plt.xticks(rotation=45)
for i, v in enumerate(country_sales.values):
    plt.text(i, v + 200, f'${v:,.0f}', ha='center')

plt.subplot(1, 3, 2)
product_sales = data.groupby('Product')['Amount ($)'].sum().sort_values(ascending=False)
colors = ['lightcoral', 'lightgreen', 'gold']
plt.bar(range(len(product_sales)), product_sales.values, color=colors)
plt.title('Total Sales Amount by Product', fontsize=14, fontweight='bold')
plt.xlabel('Product')
plt.ylabel('Amount ($)')
plt.xticks(range(len(product_sales)), product_sales.index, rotation=45, ha='right')
for i, v in enumerate(product_sales.values):
    plt.text(i, v + 200, f'${v:,.0f}', ha='center')

plt.subplot(1, 3, 3)
person_sales = data.groupby('Sales Person')['Amount ($)'].sum().sort_values(ascending=False)
plt.bar(person_sales.index, person_sales.values, color='lightsteelblue')
plt.title('Total Sales Amount by Sales Person', fontsize=14, fontweight='bold')
plt.xlabel('Sales Person')
plt.ylabel('Amount ($)')
plt.xticks(rotation=45, ha='right')
for i, v in enumerate(person_sales.values):
    plt.text(i, v + 200, f'${v:,.0f}', ha='center')

plt.tight_layout()
plt.show()
```

```python
country_boxes = data.groupby('Country')['Boxes Shipped'].sum().sort_values(ascending=False)
plt.bar(country_boxes.index, country_boxes.values, color='orange', alpha=0.7)
plt.title('Total Boxes Shipped by Country', fontsize=14, fontweight='bold')
plt.xlabel('Country')
plt.ylabel('Boxes Shipped')
plt.xticks(rotation=45)
for i, v in enumerate(country_boxes.values):
    plt.text(i, v + 10, f'{v}', ha='center')
```

```python
df_sorted = data.sort_values('Date')
plt.plot(df_sorted['Date'], df_sorted['Amount ($)'], marker='o', linewidth=2, markersize=8, color='darkgreen')
plt.title('Sales Timeline', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Amount ($)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
```

```python
plt.subplot(3, 3, 6)
plt.pie(product_sales.values, labels=product_sales.index, autopct='%1.1f%%', 
        colors=['lightcoral', 'lightgreen', 'gold'], startangle=90)
plt.title('Sales Distribution by Product', fontsize=14, fontweight='bold')
```

```python
plt.hist(data['Boxes Shipped'], bins=10, alpha=0.7, color='brown', edgecolor='black')
plt.axvline(data['Boxes Shipped'].mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {data["Boxes Shipped"].mean():.0f}')
plt.axvline(data['Boxes Shipped'].median(), color='green', linestyle='--', linewidth=2, label=f'Median: {data["Boxes Shipped"].median():.0f}')
plt.title('Distribution of Boxes Shipped', fontsize=14, fontweight='bold')
plt.xlabel('Boxes Shipped')
plt.ylabel('Frequency')
plt.legend()
```

```python
data_numeric = data.copy()
data_numeric['Country_encoded'] = pd.Categorical(data['Country']).codes
data_numeric['Product_encoded'] = pd.Categorical(data['Product']).codes
data_numeric['SalesPerson_encoded'] = pd.Categorical(data['Sales Person']).codes
data_numeric['Date_ordinal'] = pd.Categorical(data['Date']).codes

numerical_columns = ['Amount ($)', 'Boxes Shipped', 'Country_encoded', 'Product_encoded', 'SalesPerson_encoded', 'Date_ordinal']
data_analysis = data_numeric[numerical_columns]
```

```python
correlation_matrix = data_analysis.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, fmt='.3f', cbar_kws={"shrink": .8})
plt.title('Correlation Matrix', fontsize=14, fontweight='bold')
```
